```
import os

def display_main():
    print("_Quiz")
    print("\n\n Topic: General Knowledge")
    print("\n\n Questions: 20")
    print("\n\n Negative marks: 1 mark")
    print("\n Unattempted: 0 marks")
    print("\n Correct: 4 marks")
    print("\n\n Are you excited?")

def display_instructions():
    print("\nPlease go through the below instructions")
    print("\n Enter:\n  'S' for submission")
    print("  'U' for unattempted")
    print("  There are four options: a, b, c, d")

def display_read():
    print("\n\n If yes, enter 'Yes', otherwise enter 'No'")
    print("\n\n Enter your choice:")

def evaluate(answers):
    key = ['b', 'd', 'c', 'd', 'a', 'c', 'b', 'a', 'c', 'd', 'd', 'd', 'b', 'c', 'd', 'c', 'b', 'a', 'd', 'b']
    correct, wrong, unattempted = 0, 0, 0
    total_score = 0

    for i in range(len(key)):
        if answers[i] in ['U', 'n', 's', '']:
            unattempted += 1
        elif answers[i] == key[i]:
            correct += 1
            total_score += 4
        else:
            wrong += 1
            total_score -= 1

    return total_score, correct, wrong, unattempted

def review(total):
    print("\n\nREVIEW:")
    if total >= 60:
        print("\nOUTSTANDING PERFORMANCE. Keep it up")
    elif total >= 45:
        print("\nExcellent. Keep going")
    elif total >= 30:
        print("\nGOOD. Needs improvement")
    elif total >= 15:
        print("\nPOOR. Work hard")
    elif total >= 5:
        print("\nVERY BAD. Better luck next time.")
    else:
        print("\nWORST PERFORMANCE. Need more efforts.")

def show_key():
    key = ['b', 'd', 'c', 'd', 'a', 'c', 'b', 'a', 'c', 'd', 'd', 'd', 'b', 'c', 'd', 'c', 'b', 'a', 'd', 'b']
    print("\nThe Key is:")
    for i, answer in enumerate(key):
        print(f"{i + 1}. {answer}")

def main():
    name = input("Enter your name: ")
    print(f"\n    A Warm Welcome to you, {name}")

    with open("file1.txt", "w") as file:
        file.write(f"Name: {name}\n")

    display_main()
    display_read()
    choice = input().strip().lower()

    if choice != "yes":
        display_instructions()
        display_read()
        choice = input().strip().lower()

    if choice == "yes":
        answers = []
        questions = [
            "Who is the Father of Computer?\na) Dennis Ritchie\tb) Charles Babbage\nc) Adolph\td) Van Son",
            "Who is the Father of C language?\na) Dr.P.Prawin\tb) Charles Babbage\nc) Adolph\td) Dennis Ritchie",
            "What does BMW stand for?\na) Brenlie Mechanical Works\tb) Bravarian Mechanical Works\nc) Bravarian Motor Works\td) Brenlie Motor Works",
            "What is the smallest country in the world?\na) Georgia\tb) Singapore\nc) San Francisco\td) Vatican City",
            "The speed of the computer mouse is measured in?\na) Mickeys\tb) Pixels\nc) Meter per sec\td) Cursor per sec",
            "The fear of dogs is termed as?\na) Aquaphobia\tb) Acrophobia\nc) Cynophobia\td) Hypermetrophobia",
            "What is the Indian cinema that won OSCAR awards?\na) Salaar\tb) RRR\nc) Bahubali\td) KGF",
            "Capital of Russia?\na) Moscow\tb) Paris\nc) Rome\td) Greece",
            "What planets literally rain DIAMONDS?\na) Venus and Mercury\tb) Earth\nc) Saturn and Jupiter\td) Uranus and Neptune",
            "Which country has the fastest internet in the world?\na) India\tb) China\nc) United States\td) South Korea",
            "Which Indian city is also known as The City of Prime Ministers?\na) Ahmedabad\tb) New Delhi\nc) Patna\td) Allahabad",
            "Which country has the highest Life Expectancy?\na) Austria\tb) Antarctica\nc) Holland\td) Hong Kong",
            "Which hydrocarbon is termed as Fruit Ripening Gas?\na) Ethane\tb) Ethylene\nc) Methane\td) Propane",
            "Who appoints the Chief Minister?\na) President\tb) Prime Minister\nc) Governor\td) Chief Justice",
            "The largest part of the human brain is?\na) Cerebellum\tb) Midbrain\nc) Medulla Oblongata\td) Cerebrum",
            "The Gupta King who assumed the title of Vikramaditya was?\na) Skandagupta\tb) Kumaragupta\nc) Chandragupta 2\td) Samudragupta",
            "Which gas is used in the refrigerator?\na) Oxygen\tb) Freon\nc) Carbon\td) Nitrogen",
            "Which is called the METAL OF THE FUTURE?\na) Titanium\tb) Lead\nc) Platinum\td) Mercury",
            "Who opened the first school for girls in India?\na) Sarojini Naidu\tb) Indira Gandhi\nc) Lakshmi Sahgal\td) Savitribai Phule",
            "Which state in India is named as Land of Seven Languages?\na) Tamil Nadu\tb) Kerala\nc) Mumbai\td) Rajasthan",
        ]

        for i, question in enumerate(questions):
            print(f"\nQuestion {i + 1}:\n{question}\n\nEnter your choice:")
            answer = input().strip().lower()
            if answer == 's':
                break
            answers.append(answer)

        total_score, correct, wrong, unattempted = evaluate(answers)

        with open("file1.txt", "a") as file:
            file.write(f"\nTOTAL QUESTIONS: 20\nCorrect: {correct}\nWrong: {wrong}\nUnattempted: {unattempted}\nScore: {total_score}\n")

        print(f"\nTOTAL SCORE: {total_score}")
        review(total_score)

        print("\nDo you want to check the key? (yes/no):")
        if input().strip().lower() == "yes":
            show_key()

    print("\nThank you!")

if __name__ == "__main__":
    main()
```
